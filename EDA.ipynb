```python
import pandas as pd
df=pd.read_csv("heart.csv")
df
df.boxplot(column=['BMI'])  
q_low = df["BMI"].quantile(0.01)
q_hi  = df["BMI"].quantile(0.99)

df_filtered = df[(df["BMI"] < q_hi) & (df["BMI"] > q_low)]
df_filtered.boxplot(column=['BMI'])
df_filtered=df_filtered[((df_filtered['BMI'] - df_filtered['BMI'].mean()) / df_filtered['BMI'].std()).abs() < 2]
df_filtered.boxplot(column=['BMI'])
df_filtered.to_csv(r'heart.csv',index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data=pd.read_csv("heart.csv")
data
heart=data
heart["Smoking"].replace(['Yes','No'],[1,0],inplace=True)
heart["AlcoholDrinking"].replace(['Yes','No'],[1,0],inplace=True)
heart["Stroke"].replace(['Yes','No'],[1,0],inplace=True)
heart["DiffWalking"].replace(['Yes','No'],[1,0],inplace=True)
heart["PhysicalActivity"].replace(['Yes','No'],[1,0],inplace=True)
heart["Asthma"].replace(['Yes','No'],[1,0],inplace=True)
heart["KidneyDisease"].replace(['Yes','No'],[1,0],inplace=True)
heart["SkinCancer"].replace(['Yes','No'],[1,0],inplace=True)
heart["Diabetic"].replace(['Yes','No','Borderline Diabetes'],[1,0,0],inplace=True)
heart["Sex"].replace(['Male','Female'],[1,0],inplace=True)
heart["HeartDisease"].replace(['Yes','No'],[1,0],inplace=True)
matr=heart.corr(method='pearson')
plt.figure(figsize=(20,12))
plot=sns.heatmap(matr,annot=True,color='red')
```

```python
dummy=pd.get_dummies(heart["AgeCategory"])
heart1=pd.concat([heart,dummy],axis=1)
heart1=heart1.drop("AgeCategory",axis=1)
heart1
```

```python
###Train Data
x=heart1.drop("HeartDisease",axis=1)
y=heart1["HeartDisease"]
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=0)
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("the accuracy score of model is ",accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
matr=confusion_matrix(y_test,y_pred)
matr
```

```python
from mlxtend.plotting import plot_confusion_matrix
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(20,20))
plot_confusion_matrix(matr)
plt.title("Confusion Matrix")
plt.show()
```
